```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

input_folder = "GBS_score/sri_threshold_20_radius_0.05_n-splits_8_lag_0.005_scale_0.005"
metric = "sri-lag"  # ["pnp_ssi", "rp_ssi", "sri-lag", "sri-scale", "sri-dir"]

# Create a dictionary to store the data for each dataset
dataset_dict = {}

# Read and process the CSV files
for filename in os.listdir(input_folder):
    if filename.endswith(".csv"):  # only CSV files
        input_file_path = os.path.join(input_folder, filename)
        dataset, model = filename.split("_", 1)
        model, _ = model.split("_", 1)

        # Load the data and filter by the chosen metric
        df = pd.read_csv(input_file_path)
        if metric in df.columns:
            if dataset not in dataset_dict:
                dataset_dict[dataset] = []
            dataset_dict[dataset].append((model, df[metric]))

# Plot the distribution for each dataset
for dataset, models in dataset_dict.items():
    plt.figure(figsize=(10, 6))

    # Create a dataframe for plotting
    plot_data = []
    for model, values in models:
        plot_data.extend([(dataset, model, value) for value in values])

    plot_df = pd.DataFrame(plot_data, columns=["Dataset", "Model", metric])

    # Create the boxplot with seaborn
    sns.boxplot(data=plot_df, x="Model", y=metric,
                hue="Model", showfliers=False)

    # Dynamically adjust y-axis range with a smaller margin
    y_max = 1  # Add 10% of the range
    y_min = plot_df[metric].min() - 0.05 * (plot_df[metric].max() -
                                            plot_df[metric].min())  # Subtract 5% of the range
    plt.ylim(y_min, y_max)

    annotation_y = 0.5
    # Add 95% CI annotations
    for i, model in enumerate(plot_df["Model"].unique()):
        model_data = plot_df[plot_df["Model"] == model][metric]
        if not model_data.empty:
            lower, upper = np.percentile(model_data, [2.5, 97.5])  # 95% CI
            plt.text(i, annotation_y, f'95% CI: {lower:.2f} - {upper:.2f}',
                     ha='center', va='bottom', color='red', fontsize=10)

    plt.title(f"Distribution of {metric} by Model in {dataset} Dataset")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.legend(title="Model", loc="best")
    plt.show()
```

```python
import os
import shutil

# Define the base directory
base_dir = "/home/myid/nw33659/GeoBiasScore/PyGBS-main/torchspatial_tables/sri"

# Define the target subdirectories
target_dirs = {
    'dr': os.path.join(base_dir, 'dr'),
    'sc': os.path.join(base_dir, 'scale'),
    'lag': os.path.join(base_dir, 'lag')
}

# Create the target subdirectories if they don't exist
for target in target_dirs.values():
    if not os.path.exists(target):
        os.makedirs(target)

# Iterate over the files in the base directory
for filename in os.listdir(base_dir):
    if filename.endswith(".csv"):  # Assuming you are working with .csv files
        # Split the filename by '_'
        parts = filename.split('_')

        if len(parts) >= 2:
            second_last_part = parts[-2]
            last_part = parts[-1].split('.')[0]  # Remove file extension

            # Determine the target directory based on the last part of the filename
            if second_last_part == 'dr':
                target_dir = target_dirs['dr']
            elif second_last_part == 'sc':
                target_dir = target_dirs['sc']
            elif second_last_part == 'lag':
                target_dir = target_dirs['lag']
            else:
                continue  # If no match, skip the file

            # Move the file to the corresponding directory
            source_file = os.path.join(base_dir, filename)
            target_file = os.path.join(target_dir, filename)
            shutil.move(source_file, target_file)
            print(f"Moved: {filename} -> {target_dir}")
```
